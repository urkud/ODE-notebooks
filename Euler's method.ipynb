# Euler's method

Use `Ctrl+Enter` to run. Use `math.sin()`, `math.cos()`, `math.exp()`, `math.log()` for math functions. Indentations are important. `X` and `Y` are the lists of all intermediate points in the Euler's method.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
s = 0.01
x, y = 0, 1
X = [x]
Y = [y]
for i in range(1000):
    y = y + s * y
    x = x + s
    X.append(x)
    Y.append(y)
plt.plot(X, Y)
X[100], Y[100]
```
